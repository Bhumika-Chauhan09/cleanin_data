```python
import pandas as pd
import matplotlib.pyplot as plt  # Import the matplotlib.pyplot module



df = pd.read_csv('ExerciseData.csv')

# Display the first few rows of the DataFrame
#df.head()
df['distance']=pd.to_numeric(df['distance'],errors='coerce')
df['calories']=pd.to_numeric(df['calories'],errors='coerce')

df['distance'].fillna(method='ffill',inplace=True)
df['calories'].fillna(method='ffill',inplace=True)
plt.xlabel('distance')
plt.ylabel('calories')

plt.tight_layout()
plt.show()

#compute stats
mean_distance=df['distance'].mean()
median_calories=df['calories'].median()
std_dev_distance=df['distance'].std()
print (f"Mean distance:{mean_distance}")
print(f"Median calories:{median_calories}")
print(f"standard deviation of distance:{std_dev_distance}")



# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['distance'], df['calories'], alpha=0.5)
plt.xlabel('distance')
plt.ylabel('calories')
plt.title('Scatter Plot of Distance vs. Calories')
plt.show()
```
