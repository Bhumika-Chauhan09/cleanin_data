```python
import pandas as pd
import matplotlib.pyplot as plt  # Import the matplotlib.pyplot module



df = pd.read_csv('ExerciseData.csv')

# Display the first few rows of the DataFrame
#df.head()
df['distance']=pd.to_numeric(df['distance'],errors='coerce')
df['speed']=pd.to_numeric(df['speed'],errors='coerce')
df['heart_rate']=pd.to_numeric(df['heart_rate'],errors='coerce')

df['distance'].fillna(method='ffill',inplace=True)
df['speed'].fillna(method='ffill',inplace=True)
plt.xlabel('distance')
plt.ylabel('speed')

plt.tight_layout()
plt.show()

#compute stats
mean_distance=df['distance'].mean()
median_speed=df['speed'].median()
std_dev_distance=df['speed'].std()
print (f"Mean distance:{mean_distance}")
print(f"Median speed:{median_speed}")
print(f"standard deviation of distance:{std_dev_distance}")



# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['distance'], df['speed'], alpha=0.5)
plt.xlabel('distance')
plt.ylabel('speed')
plt.title('Scatter Plot of Distance vs. speed')
plt.show()
```
