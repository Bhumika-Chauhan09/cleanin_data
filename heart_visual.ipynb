```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the heart rate data from your CSV file (replace 'heart_rate_data.csv' with your file path)
csv_file = 'heart_rate_data.csv'
df = pd.read_csv(csv_file)

# Convert the 'Timestamp' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Sort the data by timestamp (if it's not already sorted)
df.sort_values(by='time', inplace=True)

# Plot the heart rate data
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['data values'], 'b-', label='Heart Rate Reading', markersize=8)
plt.xlabel('time')
plt.ylabel('data values')
plt.legend()
plt.title('Heart Rate Data Visualization')
plt.xticks(rotation=45)
plt.grid(True)



# Show the plot
plt.tight_layout()
plt.show()
```
