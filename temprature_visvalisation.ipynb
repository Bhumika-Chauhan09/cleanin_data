```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the temperature sensor data from your CSV file (replace 'temperature_data.csv' with your file path)
csv_file = 'temp_data.csv'
df = pd.read_csv(csv_file)

# Convert the 'Timestamp' column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Sort the data by timestamp (if it's not already sorted)
df.sort_values(by='Timestamp', inplace=True)

# Create a more accurate temperature graph
plt.figure(figsize=(12, 6))

# Plot the temperature data as a line chart
plt.plot(df['Timestamp'], df['Temperature(celsius)'], 'r-', label='Temperature Reading (C)', linewidth=1)

# Customize the appearance of the plot
plt.xlabel('Timestamp')
plt.ylabel('Temperature(celsius)')
plt.title('Temperature Data Visualization')
plt.grid(True)

# Format the timestamp labels on the X-axis for better readability
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))
ax.xaxis.set_major_locator(mdates.HourLocator(interval=1))  # Adjust interval as needed
plt.xticks(rotation=45)

# Ensure that the timestamp labels fit within the figure
plt.tight_layout()

# Show the plot
plt.legend()
plt.show()
```
